```python
# Кейс 1
import matplotlib.pyplot as plt
import numpy as np

# Данные, сумма которых превышает 100%
categories = ["Маркетинг", "Операционные расходы", "Зарплаты", "Развитие"]
percentages = [53, 41, 38, 18]  # В сумме 138%

def custom_pie(percentages, categories):
    total = sum(percentages)
    fig, ax = plt.subplots(figsize=(7, 7))
    wedges, texts, autotexts = ax.pie(
        percentages, labels=categories, autopct=lambda p: f'{p * total / 100:.1f}%', startangle=90,
        colors=["red", "blue", "green", "orange"], explode=[0.05] * len(categories), wedgeprops={'edgecolor': 'black'}
    )
    plt.title("Распределение бюджета компании")
    plt.show()

# Вызов функции для отрисовки диаграммы
custom_pie(percentages, categories)
```

```python
# Кейс 2
import matplotlib.pyplot as plt

# Данные о продажах
months = ["Январь", "Февраль"]
sales = [1000, 1050]  # Рост всего 5%

# Гистограмма с некорректным масштабом оси Y
plt.figure(figsize=(6, 4))
plt.bar(months, sales, color=["red", "blue"])
plt.ylim(990, 1070)  # Обрезанная ось Y
plt.title("Рост продаж")
plt.ylabel("Продажи")
plt.show()
```

```python
# Кейс 3
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Данные о росте клиентов
categories = ["2022", "2023", "2024"]
values = [1000, 1200, 1300]  # Рост клиентов
x = np.arange(len(categories))

# Создание 3D-графика
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.bar(x, values, zs=0, zdir='y', alpha=0.8, color=['red', 'blue', 'green'])

# Настройка осей
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=12)
ax.set_yticks([0])
ax.set_yticklabels(["Категории"], fontsize=12)
ax.set_zlabel("Количество клиентов", fontsize=12)
ax.set_xlabel("Годы", fontsize=12)
ax.set_title("Рост клиентов", fontsize=14)

# Показать график
plt.show()
```

```python
# Когнитивная нагрузка
import matplotlib.pyplot as plt
import numpy as np

# Данные: продажи товаров в интернет-магазине
months = ["Январь", "Февраль", "Март", "Апрель", "Май"]
sales = [100, 150, 90, 130, 110]  # Продажи в тысячах рублей

# Создание сверхперегруженного графика
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(months, sales, marker='o', linestyle='-', linewidth=3, color='blue', label='Продажи по месяцам')
ax.bar(months, sales, color='cyan', alpha=0.5, label='Сравнение с прошлым годом')
ax.scatter(months, sales, color='red', s=200, label='Пиковые значения')
ax.step(months, sales, color='purple', linestyle='dashed', linewidth=2, label='Кумулятивные продажи')
ax.fill_between(months, sales, color='yellow', alpha=0.3, label='Диапазон прогнозных значений')

# Лишние элементы
ax.set_facecolor("lightgray")  # Фон
ax.grid(True, linestyle='--', alpha=0.9, color='black')  # Усиленная сетка
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.legend(fontsize=12, loc='upper left', shadow=True, fancybox=True, framealpha=1, borderpad=1)
ax.set_title("Анализ продаж интернет-магазина", fontsize=16, fontweight='bold', color='darkred')
ax.set_xlabel("Месяцы", fontsize=14, fontweight='bold', color='darkblue')
ax.set_ylabel("Продажи (тыс. рублей)", fontsize=14, fontweight='bold', color='darkblue')

# Аннотации с дополнительными эффектами
ax.annotate("Снижение спроса в марте", xy=(2, sales[2]), xytext=(0.5, 180),
            arrowprops=dict(facecolor='black', arrowstyle='fancy', linewidth=2), fontsize=14, color='red', fontweight='bold')
ax.annotate("Рост продаж после рекламной кампании", xy=(3, sales[3]), xytext=(2.5, 190),
            arrowprops=dict(facecolor='green', arrowstyle='wedge', linewidth=2), fontsize=14, color='green', fontweight='bold')

# Дополнительные линии и разметка
ax.axhline(y=100, color='black', linestyle='dashdot', linewidth=2, label='Средний уровень продаж')
ax.axvline(x=2, color='purple', linestyle='dotted', linewidth=2, label='Критический месяц')

# Показать график
plt.show()
```

```python
# Кейс: Анализ продаж после маркетинговой кампании
# Компания запустила рекламную кампанию, ожидая роста продаж, и хочет наглядно представить результаты.

# Данные: продажи товаров в интернет-магазине
months = ["Январь", "Февраль", "Март", "Апрель", "Май"]
sales = [100, 150, 90, 130, 110]  # Продажи в тысячах рублей

# Создание перегруженного "красивого" графика
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(months, sales, marker='o', linestyle='-', linewidth=3, color='magenta', label='Продажи по месяцам')
ax.bar(months, sales, color=['red', 'yellow', 'green', 'blue', 'purple'], alpha=0.5, label='Цветные столбцы')
ax.scatter(months, sales, color='gold', s=300, label='Декоративные маркеры', edgecolors='black', linewidth=2)
ax.fill_between(months, sales, color='cyan', alpha=0.3, label='Фоновая заливка')

# Лишние декоративные элементы
ax.set_facecolor("lavender")  # Красочный фон
ax.grid(True, linestyle='-.', alpha=0.8, color='black')  # Яркая сетка
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.legend(fontsize=12, loc='upper left', shadow=True, fancybox=True, framealpha=1, borderpad=1)
ax.set_title("Как не стоит визуализировать данные: перегруженный график", fontsize=16, fontweight='bold', color='darkred')
ax.set_xlabel("Месяцы", fontsize=14, fontweight='bold', color='darkblue')
ax.set_ylabel("Продажи (тыс. рублей)", fontsize=14, fontweight='bold', color='darkblue')

# Аннотация: снижение продаж в марте
ax.annotate("Продажи снизились после акции", xy=(2, sales[2]), xytext=(0.5, 180),
            arrowprops=dict(facecolor='black', arrowstyle='fancy', linewidth=2), fontsize=14, color='red', fontweight='bold')

# Аннотация: неудачная кампания
ax.annotate("Ожидаемый рост не оправдался", xy=(4, sales[4]), xytext=(3.5, 200),
            arrowprops=dict(facecolor='green', arrowstyle='wedge', linewidth=2), fontsize=14, color='green', fontweight='bold')

# Показать график
plt.show()
```

Задание 1: Визуализация зависимости между возрастом и доходом сотрудников

Описание:
Проанализируйте, как возраст влияет на зарплату. Существует ли тренд роста доходов с возрастом?

Что можно использовать?

1. Scatter plot (Диаграмма рассеяния) → для оценки общей зависимости.

2. Boxplot (Ящик с усами) → чтобы выявить разброс доходов в разных возрастных группах.

3. Line Chart (Линейный график) → если сгруппировать данные по возрастным категориям (например, 20-30, 30-40 лет).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Загрузка данных
df = pd.read_csv("hr_data.csv")
```

```python
# Scatter Plot (Диаграмма рассеяния)
plt.figure(figsize=(10, 5)) # Создаем новую фигуру (график) размером 10x5 дюймов.
sns.scatterplot(x=df["Age"], y=df["MonthlyIncome"], alpha=0.7) # Строим диаграмму рассеяния (Scatter Plot) (alpha – полупрозначность)
plt.xlabel("Возраст")
plt.ylabel("Ежемесячный доход")
plt.title("Зависимость между возрастом и доходом")
plt.grid(True) # Добавляем сетку на график
plt.show()
```

```python
# Boxplot (Ящик с усами)
plt.figure(figsize=(10, 5))
sns.boxplot(x=pd.cut(df["Age"], bins=[20, 30, 40, 50, 60], labels=["20-30", "30-40", "40-50", "50-60"]),
            y=df["MonthlyIncome"]) # Группировка возраста по категориям
plt.xlabel("Возрастная группа")
plt.ylabel("Ежемесячный доход")
plt.title("Распределение дохода по возрастным группам")
plt.grid(True) # Добавляем сетку
plt.show()
```

```python
# Line Chart (Линейный график средних доходов по возрастным категориям)
df["AgeGroup"] = pd.cut(df["Age"], bins=[20, 30, 40, 50, 60], labels=["20-30", "30-40", "40-50", "50-60"])
# Создаем новый столбец "AgeGroup", группируя сотрудников по возрастным категориям
age_income = df.groupby("AgeGroup")["MonthlyIncome"].mean()
# Группируем данные по "AgeGroup" и вычисляет средний доход для каждой группы.

plt.figure(figsize=(10, 5))
age_income.plot(kind="line", marker="o", color="blue")
# Строим линейный график (Line Chart) средних зарплат по возрастным группам.
#   kind="line" – указываем, что строится линейный график.
#   marker="o" – добавляем точки (o), чтобы значения на линии были четко видны.
#   color="blue" – цвет линии синий.
plt.xlabel("Возрастная группа")
plt.ylabel("Средний доход")
plt.title("Средний доход по возрастным категориям")
plt.grid(True)
plt.show()
```
